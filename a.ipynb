```python
'''
Created on May 26, 2020

@author: mizuno
'''
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#df0 = pd.read_csv('./csv/int0_MCP(Ch1)_diff.csv', engine='python', header = None)
#df1 = pd.read_csv('./csv/int+_MCP(Ch1)_diff.csv', engine='python', header = None)
#df2 = pd.read_csv('./csv/int2+_MCP(Ch1)_diff.csv', engine='python', header = None)

t = 0.0
h = 0.01

# 変数の初期設定
Cfe0 = 1
Cfe1 = 1
Cfe2 = 1
ne = 1
nh = 1
k01 = -1
k10 = -1
k12 = 1
k21 = 1

def getCfe0(Cfe0, Cfe1):
    return (-k01 * Cfe0 * nh + k10 * Cfe1 * ne)

def getCfe1(Cfe0, Cfe1, Cfe2):
    return (k01 * Cfe0 * nh - (k10 * ne + k12 * nh) * Cfe1 + k21 * Cfe2 * ne)

def getCfe2(Cfe1, Cfe2):
    return (k12 * Cfe1 * nh - k21 * Cfe2 * ne)

time = []
l_Cfe0 = []
l_Cfe1 = []
l_Cfe2 = []
time.append(0)

Cfe0_tmp = 1
Cfe1_tmp = 1
Cfe2_tmp = 1
l_Cfe0.append(Cfe0_tmp)
l_Cfe1.append(Cfe1_tmp)
l_Cfe2.append(Cfe2_tmp)

while t < 10:
    Cfe0_tmp += getCfe0(Cfe0, Cfe1) * h
    Cfe1_tmp += getCfe1(Cfe0_tmp, Cfe1, Cfe2) * h
    Cfe2_tmp += getCfe2(Cfe1, Cfe2) * h
    #まとめて更新
    Cfe0 = Cfe0_tmp
    Cfe1 = Cfe1_tmp
    Cfe2 = Cfe2_tmp
    #表示
    print("t = " +str(t) )
    print("Cfe0 = "+str(Cfe0))
    print("Cfe1 = "+str(Cfe1))
    print("Cfe2 = "+str(Cfe2))
    t += h
    time.append(t)
    l_Cfe0.append(Cfe0)
    l_Cfe1.append(Cfe1)
    l_Cfe2.append(Cfe2)

x = [100, 200, 300, 400, 500, 600]
y1 = [10, 20, 30, 50, 80, 130]
y2 = [10, 15, 30, 45, 60, 75]

plt.plot(time, l_Cfe0, label="CFe0")
plt.plot(time, l_Cfe1, label="CFe1")
plt.plot(time, l_Cfe2, label="CFe2")
plt.legend()
#plt.plot(x, y1, marker="o", color = "red", linestyle = "--")
#plt.plot(x, y2, marker="v", color = "blue", linestyle = ":")
plt.grid()
plt.show()

#print(df0[0][0])
#print(df1[0][0])
#print(df2[0][0])
#目的関数
#P = (df0[0][0] - Cfe0)**2 + (df1[0][0] - Cfe1)**2 + (df2[0][0] - Cfe2)**2
#P /= 2
#print('目的関数値')
#print(P)
```
